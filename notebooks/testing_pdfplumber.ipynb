```python
import pdfplumber
```

```python
with pdfplumber.open("../pdf_files/owner_manual_p283-p300.pdf") as pdf:
    for page_number, page in enumerate(pdf.pages, start=1):
        print(f'page number: {page_number}')
        page = page.crop((0, 0.25*page.height, page.width, 0.77*page.height))
        im = page.to_image()
        im.draw_rects(page.extract_words())
        im.save(f"image{page_number}.png", format="PNG", quantize=True, colors=256, bits=8)
```
